```python
import seaborn as sns
import pandas as pd
```

```python
flight = sns.load_dataset("flights")
```

```python
flight
```

```python
fp = flight.pivot_table(values="passengers",   index="year", columns="month", aggfunc="sum")
```

```python
fp
```

```python
sns.heatmap(data=fp) # Heat map depicts intensity of each values in a dataframe in colorful way 
```

```python
sns.heatmap(data = fp, annot = True)
```

```python
sns.heatmap(data = fp, annot = True, linewidths=0.5)
```

```python
sns.heatmap(data = fp, annot = True, fmt = "d") # fmt = "d" for numbers, "s" for strings
```

```python
sns.heatmap(data = fp, annot = True, fmt = "d", linewidths=0.5) # fmt = "d" for numbers, "s" for strings
```

```python
sns.heatmap(data = fp, annot = True, fmt = "d", linewidths=0.5, cmap="summer")  # cmap is used to chnage the color code
```

```python
sns.heatmap(data = fp, annot = True, fmt = "d", linewidths=0.5, cmap = "winter")  
```

```python
sns.heatmap(data = fp, annot = True, fmt = "d", linewidths=0.5, cmap = "viridis")  
```
