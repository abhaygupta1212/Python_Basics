## Important Point

In a data sets there are various features (f1, f2, f3, f4...)

Step 1: Divide it into independent and dependent feature.
        Dependent Feature is the output feature which we need to compute (forcast)
        
Step 2: If we have only two features, we can draw a 2D graph and see the relationship between them.
        If we have 3 features, then we can draw a 3D diagaram to see the relationship
        If we have more than 3 features, it is not possible to draw diagram to see the relationship

Univariant Analysis: When we have only 1 feature

Bivarient Analysis: When we have 2 features [here we draw 2D diagram]

### In seaborn we have multiple plot
#### 1. JoinPlot: Help us to do BiVarient Analysis
#### 2. PairPlot: Multivariate Analysis
#### 3. Univariate Analysis 
        i. HISTOGRAM --> displot
        ii. Box Plot and Whisker Plot --> boxplot

## Univariate Analysis (analysing one Attribute) = DISTPLOT (HISTOGRAM)

```python
sns.distplot(df["total_bill"], kde = False, bins = 5)
```

```python
sns.distplot(df["total_bill"], kde = True, bins = 100) # kde --> Kernal Density Estiimation (PDF)

# bins --> Number of Histogram

# Note: When kde is shown then y axis is probaility, whereas when kde is not shown the y axis is values
```

```python
sns.boxplot(df["tip"])
```

```python
sns.boxplot(data = df) # Dra box plot for all the numberic features
```

## Bivriate Analysis [2 Attribute] --> JOINPLOT --> Scatter Plot


```python
sns.jointplot(x = "total_bill", y = "tip", data=df)
# Scanner Diagram, with frequency histograme
```

```python
sns.jointplot(x = "total_bill", y = "tip", data=df, kind = "reg") 
# It shall also add a regression line and probaility density function
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = "kde")
# kde stands for Krnel Density Estimate --> It is analogous to Histogram 
# and helps to undestand the concentartion of data
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df, kind = "hex")
```

## MULTIVARIATE ANALYSIS (>2 ATTRIBUTE) --> PAIRPLOT
Here we will make all possible pairs that is possible for attribute and plot them on graph


```python
sns.pairplot(data = df)
# The pair of two same attribute leads two univariate analysis --> Histogram
# The pair of two different attribute forms bivariate analysis --> Scatter plot
```

## HUE parameter
In pair plot we have hue parameter --> 

Scatter Plot : it differntiate the poinnts by different color based on the attribute which is used in hue

Histogram (kde) : It draws multiple kde for each diffrenet value in the attribute chosen in hue

```python
sns.pairplot(data = df, hue = "day")
```
