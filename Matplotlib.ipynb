```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# %matplotlib inline 
# enable to dsiplay the graph without .show()
```

```python
x = np.arange(-100, 101, 1) 

y = x**3

```

```python
# Scatter Plot
plt.scatter(x, y)
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.title("y = x^3")
plt.savefig("data sets\\cubic graph.png")
```

```python
# Line Plot
plt.plot(x, y , "*--r", linewidth = 1, markersize = 1)
```

## Ticklable_format: used to format the x and y axis dsipalyed

```python
## style + How to display two differnet graph (use plt.show() two times)
plt.plot(x, y)
plt.show()
plt.ticklabel_format(style = "plain") # The default style is sci, so it is displayed in scientific notation
plt.plot(x,y)
plt.show()
```

```python
# How to display in a single graph
y = 2*x + 3
y_ = 3*x + 4
plt.plot(x, y, c = "r")
plt.plot(x, y_, c= "g")
plt.show()
```
